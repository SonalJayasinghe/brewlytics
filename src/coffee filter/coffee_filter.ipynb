## Identify Unnecessary Columns 

```python
df.head()
```

```python
def calculate_pre_statistics(data):
    summary = pd.DataFrame({
        'dtype': df.dtypes,
        'non_null_count': df.notnull().sum(),
        'missing_pct': (df.isnull().mean() * 100).round(2),
        'unique_values': df.nunique(dropna=True)
    })
    summary['feature_type'] = summary['dtype'].apply(lambda x: 'numeric' if np.issubdtype(x, np.number) else 'categorical')
    summary.reset_index(inplace=True)
    summary.rename(columns={'index': 'column'}, inplace=True)
    summary.sort_values(by='missing_pct', ascending=False, inplace=True)
    display(summary)

calculate_pre_statistics(df)
```

I can see that Certification.Body, Certification.Address, Certification.Contact, ICO.Number, Lot.Number, Species, Unnamed: 0  columns can be deleted safely. 

Dropped the columns in following reasons
1. Lot. Number - > 50% missing values
2. Certification.Body, Certification.Address, Certification.Contact, ICO.Number - Unnecessary columns for the project
3. Species - No varience 
4. Unnamed 0 - just an index field
5. Unit of measure - already have it as m in altitude mean meters

```python
df.drop(columns=['Certification.Body', 'Certification.Address','Certification.Contact', 'ICO.Number', 'Lot.Number', 'Species', 'Unnamed: 0', 'unit_of_measurement' ], inplace=True)
```

```python
calculate_pre_statistics(df)
```

## Checking for Outlairs


### checking the outliers for missing value columns of cupper points and mositure

```python

calculate_pre_statistics(df)
```

```python
category_cols = ['Cupper.Points','Moisture']
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(5, 5))
axes = axes.flatten()

def boxplot_drawer(axes,category_cols ):
    for ax, col in zip(axes, category_cols):
        ax.boxplot(df[col].dropna(), vert=True, whis=1.5)
        ax.set_title(col)
        ax.tick_params(axis='x', which='both', bottom=False, labelbottom=False)

    plt.tight_layout()
    plt.show()

boxplot_drawer(axes,category_cols )
```

as we can see there are some outliars. now we have to remove those

```python
for col in ['Cupper.Points', 'Moisture']:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_fence = Q1 - 1.5 * IQR
    upper_fence = Q3 + 1.5 * IQR

    mask = (df[col] < lower_fence) | (df[col] > upper_fence)

    outliers = df.loc[mask, ['Cupper.Points','Moisture', *[c for c in df.columns if c not in ['Cupper.Points','Moisture']]]]
    print(f"\n=== Outliers in '{col}' ===")
    print(outliers[[col]].sort_values(by=col))
```

```python
df.drop(index=[1310, 735, 636, 379, 352, 1484, 526], inplace=True)
```

```python
#drop moisture outliers values of 0
zero_idx = df[df['Moisture'] == 0].index
df.drop(index=zero_idx, inplace=True)
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(5, 5))
axes = axes.flatten()
boxplot_drawer(axes,category_cols )
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import gaussian_kde

fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(10, 4))
axes = axes.flatten()

for ax, col in zip(axes, category_cols):
    data = df[col].dropna().values
    ax.hist(data, bins=20, density=True, edgecolor='black', alpha=0.6)
    kde = gaussian_kde(data)
    x = np.linspace(data.min(), data.max(), 200)
    ax.plot(x, kde(x), lw=2)
    ax.set_title(col)
    ax.grid(axis='y', alpha=0.5)

plt.tight_layout()
plt.show()
```

It appeas to be slightly skewed so have to go with median to full null values for cupper points and moisture
