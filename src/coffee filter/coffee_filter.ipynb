```python
def calculate_pre_statistics(data):
    summary = pd.DataFrame({
        'dtype': df.dtypes,
        'non_null_count': df.notnull().sum(),
        'missing_pct': (df.isnull().mean() * 100).round(2),
        'unique_values': df.nunique(dropna=True)
    })
    summary['feature_type'] = summary['dtype'].apply(lambda x: 'numeric' if np.issubdtype(x, np.number) else 'categorical')
    summary.reset_index(inplace=True)
    summary.rename(columns={'index': 'column'}, inplace=True)
    summary.sort_values(by='missing_pct', ascending=False, inplace=True)
    display(summary)

calculate_pre_statistics(df)
```

## 1.1 Reusable Functions Here

```python
def draw_histogram_and_boxplot(df, col):
    fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(12, 5))
    axes = axes.flatten()

    # Histogram
    sns.histplot(df[col], bins=30, kde=True, ax=axes[0], color='orange', edgecolor='black')
    axes[0].set_title(f"Histogram of {col}")
    axes[0].set_xlabel(col)
    axes[0].set_ylabel("Frequency")

    # Boxplot
    sns.boxplot(x=df[col], ax=axes[1], color="skyblue", showfliers=True)
    axes[1].set_title(f"Boxplot of {col}")
    axes[1].set_xlabel(col)

    plt.tight_layout()
    plt.show()
plt.figure(figsize=(14, 5))

```

```python
def plot_top_categories(df, column_name, top_n=15, palette="viridis"):
    # Convert to string to avoid categorical rendering bugs
    counts = df[column_name].astype(str).value_counts().head(top_n)

    plt.figure(figsize=(12, 6))
    sns.barplot(y=counts.index, x=counts.values, palette=palette)
    plt.title(f"Top {top_n} Categories in '{column_name}'")
    plt.xlabel("Number of Samples")
    plt.ylabel(column_name)
    plt.tight_layout()
    plt.grid(axis='x', linestyle='--', alpha=0.7)
    plt.show()
```

```python
def categorical_boxplot_drawer(df, categorical_col_name, rotation=90):
    plt.figure(figsize=(12, 6))
    sns.boxplot(x=df[categorical_col_name].astype(str), y=df["Total.Cup.Points"], color="orange", showfliers=True)
    plt.xticks(rotation=rotation)
    plt.title(f"{categorical_col_name} vs Total.Cup.Points")
    plt.xlabel(categorical_col_name)
    plt.ylabel("Total.Cup.Points")
    plt.tight_layout()
    plt.show()
```

```python
outlier_features = [
    "Flavor", "Aftertaste", "Cupper.Points", "Balance", "Acidity",
    "Aroma", "Body", "Moisture",
]

outlier_summary = []

for col in outlier_features:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
    percent_outliers = 100 * len(outliers) / len(df)
    
    outlier_summary.append({
        "Feature": col,
        "Q1": round(Q1, 2),
        "Q3": round(Q3, 2),
        "Lower Bound": round(lower_bound, 2),
        "Upper Bound": round(upper_bound, 2),
        "Outlier Count": len(outliers),
        "Percent Outliers (%)": round(percent_outliers, 2)
    })

df_outlier_summary = pd.DataFrame(outlier_summary)
display(df_outlier_summary)
```

```python

for feature in outlier_features:
    Q1 = df[feature].quantile(0.25)
    Q3 = df[feature].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    df[feature] = df[feature].clip(lower=lower_bound, upper=upper_bound)

df_capped_preview = df[outlier_features].copy()
display(df_capped_preview)
```

# 7. Model Traning

```python
df_final_model_data = pd.read_csv('/Users/admin/Documents/PICT/Year 1 Sem 1/ICT/Data Mining and Warehousing/Project/Brewlytics/data/df_final_model_data.csv')
```

## Linear Regression

```python
import pandas as pd 
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score, root_mean_squared_error, mean_absolute_error, mean_absolute_percentage_error
```

```python
df_final_model_data.info()
```

```python
X = df_final_model_data.drop(columns=["Total.Cup.Points"])
y = df_final_model_data["Total.Cup.Points"]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
lr_model = LinearRegression()
lr_model.fit(X_train, y_train)
```

```python
y_pred = lr_model.predict(X_test)
```

```python
r2 = r2_score(y_test, y_pred)
rmse = root_mean_squared_error(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)
mape = mean_absolute_percentage_error(y_test, y_pred)
```

```python
{
    "Model": "Linear Regression",
    "RMSE": round(rmse, 3),
    "R2 Score": round(r2, 3),
    "MAE": round(mae, 3),
    "MAPE (%)": round(mape * 100, 2),
}
```

## Random Forest Regression

```python
from sklearn.ensemble import RandomForestRegressor
rf_model = RandomForestRegressor(random_state=42)
rf_model.fit(X_train, y_train)
rf_pred = rf_model.predict(X_test)
```

```python
rf_rmse = root_mean_squared_error(y_test, rf_pred)
rf_r2 = r2_score(y_test, rf_pred)
rf_mae = mean_absolute_error(y_test, rf_pred)
rf_mape = mean_absolute_percentage_error(y_test, rf_pred)
```

```python
{
    "Model": "Random Forest Regression",
    "RMSE": round(rf_rmse, 3),
    "R2 Score": round(rf_r2, 3),
    "MAE": round(rf_mae, 3),
    "MAPE (%)": round(rf_mape * 100, 2),
}
```

## Lasso and Ridge

```python
from sklearn.linear_model import Ridge, Lasso

ridge_model = Ridge(random_state=42)
ridge_model.fit(X_train, y_train)
ridge_pred = ridge_model.predict(X_test)

lasso_model = Lasso(random_state=42)
lasso_model.fit(X_train, y_train)
lasso_pred = lasso_model.predict(X_test)
```
